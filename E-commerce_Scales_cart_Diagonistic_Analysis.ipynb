```python
# E-commerce Scales cart Abondment Analysis
# Diagnostic Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#data
customer_name = ['Arya', 'Mani', 'Karthik', 'Vardhan', 'Kumar']
internet_speed = [3, 2.5, 10, 8, 9]
payment_failures = [1, 1, 0, 0, 0]
delivery_charges = [999, 1299, 299, 599, 499]
perchases_completed = [0, 0, 1, 1, 1]
customer_data= pd.DataFrame ({
    'customer_name': customer_name,
    'internet_speed': internet_speed,
    'payment_failures': payment_failures,
    'delivery_charges': delivery_charges,
    'perchases': perchases_completed})
print(customer_data)
#Total Abondment
print(f"\nTotal Abondent: \n{customer_data['perchases'].value_counts()}")

#Diagnosis Rules
customer_data['poor_internet_speed'] = customer_data['internet_speed'] > 5
customer_data['payment_failures'] = customer_data['payment_failures'] >0
customer_data['High_delivery_charges'] = customer_data['delivery_charges'] > 600
customer_data['perchases'] = customer_data['perchases'] > 0

#Diagnosis Table
print("\nData Diagnosis Table")
print(customer_data[['customer_name','poor_internet_speed', 'payment_failures', 'High_delivery_charges', 'perchases']])

#Root Causes
print("\nRoot Causes Analysis: ")
print("\ncustomers with less internet speed: ", customer_data['poor_internet_speed'].sum())
print("\ncustomers with payment failures: ", customer_data['payment_failures'].sum())
print("\ncustomers with High delivery charges: ", customer_data['High_delivery_charges'].sum())

plt.figure(figsize=(10,5))
plt.bar(customer_data['customer_name'], customer_data['delivery_charges'])
plt.title('Delivery Charges')
plt.xlabel('Customers')
plt.ylabel('Delivery Charges')
plt.show()
```
