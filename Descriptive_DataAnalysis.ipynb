```python
##### Descriptive Analysis #####

import matplotlib.pyplot as plt
import statistics

members_names = ['Aman', 'Riya', 'Karan', 'Sneha', 'Rahul']

ages = [22, 25, 24, 23, 26]

workout_hours = [2, 1.5, 2.5, 1, 3]

calories_burned = [500, 350, 600, 300, 750]

print(".....Fitness Dataset displayed.....")

for i in range(len(members_names)):
    print(members_names[i], ages[i], workout_hours[i], calories_burned[i])

#mean
avg_work = sum(workout_hours)/len(workout_hours)
print("\nAverage workout hours: ", avg_work)

#Maximum calories
height_calories = max(calories_burned)
print("\nMaximum calories burned: ", height_calories)

#minimum
lowest_calories = min(calories_burned)
print("\nMinimum calories burned: ", lowest_calories)

#Median
median_calories = statistics.median(calories_burned)
print("\nMedian_calories burned: ", median_calories)

#standered deviation
std_dev = statistics.stdev(calories_burned)
print("\nStandered deviation: ", std_dev)

#Most active member
index = calories_burned.index(height_calories)
print("\nMost active memeber: ", members_names[index])

plt.bar(members_names, calories_burned)
plt.xlabel("members")
plt.ylabel("calories burned")
plt.title("Fitness Analysis")
plt.show()


```
