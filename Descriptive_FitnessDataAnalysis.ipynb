```python
##### Descriptive Analysis #####

import matplotlib.pyplot as plt
import pandas as pd
import statistics

data = pd.read_csv("/content/gym_members_exercise_tracking_synthetic_data_cleaned_cleaned.csv")

data = data.dropna()
Workout_Type = data["Workout_Type"]
ages = data["Age"]
Average_Duration = data["Session_Duration (hours)"]
calories_burned = data["Calories_Burned"]

print(".....Fitness Dataset displayed.....")

#for i in range(len(Workout_Type)):
  #print(Workout_Type[i], ages[i], Average_Duration[i], calories_burned[i])

#mean
avg_weight = sum(Average_Duration)/len(Average_Duration)
print("\nAverage Weight: ", avg_weight)

#Maximum calories
height_calories = max(calories_burned)
print("\nMaximum calories burned: ", height_calories)

#minimum
lowest_calories = min(calories_burned)
print("\nMinimum calories burned: ", lowest_calories)

#Median
median_calories = statistics.median(calories_burned)
print("\nMedian_calories burned: ", median_calories)

#standered deviation
std_dev = statistics.stdev(calories_burned.tolist())
print("\nStandered deviation: ", std_dev)

#Best Workout Type
index = calories_burned.idxmax()
print("\nBest Workout Type: ", Workout_Type[index])

plt.bar(Workout_Type, calories_burned)
plt.xlabel("Workout Type")
plt.ylabel("calories burned")
plt.title("Fitness Analysis")
plt.show()

plt.scatter(Workout_Type, calories_burned)
plt.xlabel("Workout Type")
plt.ylabel("calories burned")
plt.title("Fitness Analysis")
plt.show()
```
