```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

marks = [70, 80, 90, 60, 100]
hours = [2, 3, 5, 1, 6]

print("Mean", np.mean(marks))
print("Median", np.median(marks))
print("Variance", np.var(marks))
print("Standard Deviation", np.std(marks))
print("Correlation: ", np.corrcoef(marks, hours))

data = pd.DataFrame({'Marks': marks, 'Hours': hours})
print(data)

plt.figure(figsize=(10, 5))
sns.barplot(x='Marks', y='Hours', data=data)
plt.show()

plt.figure(figsize=(10, 5))
sns.lineplot(x='Marks', y='Hours', data=data)
plt.show()

plt.figure(figsize=(10, 5))
sns.scatterplot(x='Marks', y='Hours', data = data)
plt.show()
```
